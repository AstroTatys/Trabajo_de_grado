```python
#Librerias.
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

Propiedades físicas de las galaxias escogidas:
- Magnitud absoluta
- Distancia

Estas 2 propiedades físicas se escogen ya que estas permiten hallar las relaciones de Tully-Fisher y por medio de estas encontrar las demás propiedades de interes.

propiedades galaxias espirales.PNG

# Para galaxias tipo Sa.

```python
#Magnitud absoluta    
MB_a = np.random.uniform(-23.00,-17.00,1500) #1500 galaxias

#Distancia
#Galaxia más lejana: 13.400 millones de años luz
d = np.random.uniform(10e6,4.11e9,1500) #pc
```

Por medio de la ecuación del módulo de la distancia
$m_{B} - M_{B} = 5 log_{10}(\frac{d}{10pc})$
podemos encontrar:

```python
#Magnitud aparente
mB_a = MB_a+5*np.log(10,(d/10))
```

De la relación de Tully-Fisher para galaxias de tipo Sa $M_{B} = -9.95 log_{10}(Vmax)+ 3.15$ podemos encontrar la velocidad circular máxima de rotación.

```python
#Velocidad circular máxima de rotación
Vmax_a = 10**((MB_a-3.15)/-9.95)
```

Usando la relación Masa-Luminosidad : $log_{10}(R_{25}) = -0.249 M_{B} - 4.00$

```python
#Radio 25
R25_a = 10**(-0.249*MB_a-4.00)

#Masa interior a R25
G = 6.67e-20 #km^3 kg^-1 s^-2
M_a = ((Vmax_a**2)*R25_a)/G

#Luminosidad de la galaxia
L_tot_a = 10**((MB_a-5.47)/-2.5)

#Razon Masa-Luminosidad Galaxias tipo Sa = 6.2
#Masa total de la galaxia
M_tot_a = 6.2*L_tot_a

#Luminosidad del bulbo
L_B_a = 0.3*L_tot_a

#Masa en la forma de gas
M_gas_a = 0.04*M_tot_a

#Indice de color B-V a cada galaxia.
B_V_a = 0.75
```

```python
#Gráfica Vmax vs MB
plt.title("Vmax vs Magnitud Absoluta Galaxias tipo Sa")
plt.ylabel("MB_a")
plt.xlabel("Vmax[km $s^{-1}$]")
plt.gca().invert_yaxis()
plt.plot(Vmax_a,MB_a,"*")
```

# Para las galaxias tipo Sb

```python
#Magnitud absoluta    
MB_b = np.random.uniform(-23.00,-17.00,1500)

#Distancia , galaxi mas lejana 13.400 millones de años luz
d = np.random.uniform(10e6,4.11e9,1500) #pc

#Magnitud aparente
mB_b = MB_b+5*np.log(10,(d/10))
```

De la relación de Tully-Fisher para galaxias de tipo Sb $M_{B} = -10.2 log_{10}(Vmax)+ 2.71$ podemos encontrar la velocidad circular máxima de rotación.

```python
#Velocidad circular máxima de rotación
Vmax_b = 10**((MB_b-2.71)/-10.2)

#Radio25
R25_b = 10**(-0.249*MB_b-4.00)

#Masa interior a R25
G = 6.67e-20 #km^3 kg^-1 s^-2
M_b = ((Vmax_b**2)*R25_b)/G

#Luminosidad de la galaxia
L_tot_b = 10**((MB_b-5.47)/-2.5)

#Razon Masa-Luminosidad Galaxias tipo Sb = 4.5
#Masa total de la galaxia
M_tot_b = 4.5*L_tot_b

#Luminosidad del bulbo
L_B_b = 0.13*L_tot_b

#Masa en la forma de gas
M_gas_b = 0.08*M_tot_b

#Indice de color B-V a cada galaxia.
B_V_b = 0.64
```

```python
#Gráfica Vmax vs MB
plt.title("Vmax vs Magnitud Absoluta Galaxias tipo Sb")
plt.ylabel("MB_b")
plt.xlabel("Vmax[km $s^{-1}$]")
plt.gca().invert_yaxis()
plt.plot(Vmax_b,MB_b,"*",color="red")
```

# Para las galaxias tipo Sc

```python
#Magnitud absoluta    
MB_c = np.random.uniform(-22.00,-16.00,1500)

#Distancia , galaxi mas lejana 13.400 millones de años luz
d = np.random.uniform(10e6,4.11e9,1500) #pc

#Magnitud aparente
mB_c = MB_c+5*np.log(10,(d/10))
```

De la relación de Tully-Fisher para galaxias de tipo Sc $M_{B} = -11.0 log_{10}(Vmax)+ 3.31$ podemos encontrar la velocidad circular máxima de rotación.

```python
#Velocidad circular máxima de rotación
Vmax_c = 10**((MB_c-3.31)/-11.0)

#Radio25
R25_c = 10**(-0.249*MB_c-4.00)

#Masa interior a R25
G = 6.67e-20 #km^3 kg^-1 s^-2
M_c = ((Vmax_c**2)*R25_c)/G

#Luminosidad de la galaxia
L_tot_c = 10**((MB_c-5.47)/-2.5)

#Razon Masa-Luminosidad Galaxias tipo Sc = 2.6
#Masa total de la galaxia
M_tot_c = 2.6*L_tot_c

#Luminosidad del bulbo
L_B_c = 0.05*L_tot_c

#Masa en la forma de gas
M_gas_c = 0.16*M_tot_c

#Indice de color B-V a cada galaxia.
B_V_c = 0.52
```

```python
#Gráfica Vmax vs MB
plt.title("Vmax vs Magnitud Absoluta Galaxias tipo Sc")
plt.ylabel("MB_c")
plt.xlabel("Vmax[km $s^{-1}$]")
plt.gca().invert_yaxis()
plt.plot(Vmax_c,MB_c,"*",color="purple")
```
